```python
from langchain.document_loaders import TextLoader
```

```python
loader=TextLoader("MG University.txt")
```

```python
document=loader.load()
```

```python
print(document)
```

```python
#preprocessing

import textwrap
```

```python
def wrap_text_preserve_newlines(text, width=110):
    #split the input text into lines based on newline characters
    lines = text.split("\n")
    wrapped_lines = [textwrap.fill(line, width=width) for line in lines]
    wrapped_text = "\n".join(wrapped_lines)
    
    return wrapped_text
```

```python
print(wrap_text_preserve_newlines(str(document[0])))
```

```python
#Text Splitting

from langchain.text_splitter import CharacterTextSplitter

text_splitter=CharacterTextSplitter(chunk_size=1200, chunk_overlap=0)

docs=text_splitter.split_documents(document)

print(docs[0])
print(len(docs))
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
```

```python
embeddings= HuggingFaceEmbeddings()
```

```python
db = FAISS.from_documents(docs, embeddings)
```

```python
query = "Mg university"

doc=db.similarity_search(query)
```

```python
print(wrap_text_preserve_newlines(str(doc[0].page_content)))
```
